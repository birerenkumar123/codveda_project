```python
def emperial(x):
    lst=[]
    ll=np.mean(x)-3*np.std(x)
    ul=np.mean(x)+3*np.std(x)
    for i in x:
        if i>=ll and i<=ul:
            lst.append(i)
    return lst

sns.distplot(emperial(data1['open']))
```
